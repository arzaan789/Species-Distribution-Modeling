```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

# Load UK countries shapefile (EPSG:27700)
uk_countries = gpd.read_file('data/CTRY_DEC_2024_UK_BFC.shp')
uk_countries = uk_countries[uk_countries['CTRY24NM'].isin(['England', 'Scotland', 'Wales', 'Northern Ireland'])]

# Load your point data
data = pd.read_csv("bat_final_data_preprocessed_with_coords.csv")

# Convert points to GeoDataFrame with lat/lon CRS
points_gdf = gpd.GeoDataFrame(
    data,
    geometry=gpd.points_from_xy(data.decimalLongitude, data.decimalLatitude),
    crs='EPSG:4326'
)

# Reproject points to British National Grid (EPSG:27700)
points_gdf = points_gdf.to_crs('EPSG:27700')

# Filter points by type
presence_points = points_gdf[points_gdf['type'] == 'presence']
background_points = points_gdf[points_gdf['type'] == 'background']

# Plot side-by-side maps
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 12), sharey=True)

# Plot UK boundaries on both axes
uk_countries.boundary.plot(ax=ax1, color='black', linewidth=1.5)
uk_countries.boundary.plot(ax=ax2, color='black', linewidth=1.5)

# Plot presence points
presence_points.plot(ax=ax1, color='red', markersize=30)
ax1.set_title('Presence Points', fontsize=30)
ax1.set_axis_off()

# Plot background points
background_points.plot(ax=ax2, color='blue', markersize=30)
ax2.set_title('Background Points', fontsize=30)
ax2.set_axis_off()

plt.tight_layout()
plt.show()
```
